### PROJECT AIM AND DATA
#### AIM
- the aim of the project is to analyse the given data and isolate the outliers from the data
- we will be doing this in 2 ways using isolation forest and then using one class svm algorithm
#### DATA
- we will be generating random 2d datapoints that form several clusters 
  inherently
- 
-------------------------------------------------------------[BACK TO TOP](#top)

```python
# generating 2000 datapoints and visualising them
from sklearn.datasets import make_blobs
data = make_blobs(n_samples = 2000,n_features = 2,centers=6, 
                             cluster_std = 1.8)
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

<a id = ''></a>
###  USING ISOLATION FOREST
-------------------------------------------------------------[BACK TO TOP](#top)

```python
# using isolation forest
from sklearn.ensemble import IsolationForest
aobject = IsolationForest(n_estimators=10,contamination=.1)
aobject.fit(data[0])  # fitting the data into the model
pred = aobject.predict(data[0]) # predict
plt.scatter(data[0][:,0],data[0][:,1],c=pred,cmap='rainbow')
```

<a id = ''></a>
### USING ONE CLASS SVM
-------------------------------------------------------------[BACK TO TOP](#top)

```python
from sklearn.svm import OneClassSVM
aobject2  = OneClassSVM(nu=.05,kernel='rbf',gamma='auto')
aobject2.fit(data[0])
pred = aobject2.predict(data[0])
plt.scatter(data[0][:,0],data[0][:,1],c=pred,cmap='rainbow')
# nu is percent you think are anomalies tune it to get best results
# you can even tune the kernel to get best results
```
